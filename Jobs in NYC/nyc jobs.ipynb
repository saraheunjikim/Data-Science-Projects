## My goal is to: find out how many full/part time jobs are therefore my_jobs list. 
To do so,
1) Create seperate data frame that contains only my_jobs list.
    - from "job_category", 
2) Count all full/part time.
3) Make a countplot with a new data daframe.

```python
my_jobs_df = DataFrame(data=data["job_category"])
```

```python
my_jobs_df.dropna(inplace=True)
```

```python
my_jobs_df['number_of_jobs'] = data['full-time/part-time']
```

```python
my_jobs_df.dropna(inplace=True)
```

```python
my_jobs_df.head()
```

```python
my_jobs_df = my_jobs_df.groupby(['job_category']).count()
```

```python
my_jobs_df = my_jobs_df.reset_index()
```

```python
plt.figure(figsize=(50,100))
sns.barplot(data=my_jobs_df, y='job_category', x='number_of_jobs')
```

```python
my_jobs_df.head()
```

```python
my_jobs_df = my_jobs_df[my_jobs_df["job_category"].isin(my_jobs)]
```

```python
my_jobs_df
```

```python
my_jobs_df = my_jobs_df.reset_index()
```

```python
my_jobs_df = my_jobs_df.drop('index', axis=1)
```

```python
my_jobs_df
```
