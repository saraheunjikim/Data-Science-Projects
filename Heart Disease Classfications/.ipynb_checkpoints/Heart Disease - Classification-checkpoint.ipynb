```python
df = df.drop(columns = ['cp', 'thal', 'slope'])
df.head()
```

### Creating Model for Logistic Regression

```python
y = df.target.values
x_data = df.drop(['target'], axis=1)
```

#### Normalize Data


```python
# Normalize
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train = x_train.T
y_train = y_train.T
x_test = x_test.T
y_test = y_test.T
```

```python
# initialize
def initialize(dimension):
    weight = np.full((dimension, 1), 0.01)
    bias = 0.0
    return weight, bias
```

#### Sigmoid Function


```python
def sigmoid(z):
    y_head = 1 / (1 + np.exp(-z))
    return y_head
```

#### Forward and Backward Propagation


#### Cost Function


#### Gradient Descent


```python
def forwardBackward(weight, bias, x_train, y_train):
    # Forward
    y_head = sigmoid(np.dot(weight.T, x_train) + bias)
    loss = -(y_train * np.log(y_head) + (1-y_train) * np.log(1-y_head))
    cost = np.sum(loss) / x_train.shape[1]
    
    # Backward
    derivative_weight = np.dot(x_train, ((y_head - y_train).T)) / x_train.shape[1]
    derivative_bias = np.sum(y_head - y_train) / x_train.shape[1]
    gradients = {"Derivative Weight":derivative_weight, "Derivative Bias":derivative_bias}
    
    return cost, gradients
```

```python
def update(weight, bias, x_train, y_train, learningRate, iteration):
    costList = []
    index = []
    
    # for each iteration, update weight and bias values
    for i in range(iteration):
        cost, gradietns = forwardBackward(weight, bias, x_train, y_train)
        weight = weight - learningRate * gradietns['Derivative Weight']
        bias = bias - learningRate * gradietns['Derivative Bias']
        
        costList.append(cost)
        index.append(i)
        
    parameters = {'weight':weight, 'bias':bias}
    
    print("iterations: ", iteration)
    print("cost: ", cost)
    
    plt.plot(index, costList)
    plt.xlabel('Number of Iteration')
    plt.ylabel('Cost')
    plt.show()
    
    return parameters, gradietns
```

```python
def predict(weight, bias, x_test):
    z = np.dot(weight.T, x_test) + bias
    y_head = sigmoid(z)
    
    y_prediction = np.zeros((1, x_test.shape[1]))
    
    for i in range(y_head.shape[1]):
        if y_head[0, i] <= 0.5:
            y_prediction[0, i] = 0
        else:
            y_prediction[0, i] = 1
    return y_prediction
```

```python
def logistic_regression(x_train, y_train, x_test, y_test, learningRate, iteration):
    dimension = x_train.shape[0]
    weight, bias = initialize(dimension)
    
    parameters, gradients = update(weight, bias, x_train, y_train, learningRate, iteration)
    
    y_prediction = predict(parameters['weight'], parameters['bias'], x_test)
    
    print("Manual Test Accuary: {:.2f}%".format((100 - np.mean(np.abs(y_prediction - y_test)) * 100)))
```

```python
logistic_regression(x_train, y_train, x_test, y_test, 1, 100)
```

### SKLearn Logistic Regression

```python
accuracies = {}
lr = LogisticRegression()
lr.fit(x_train.T, y_train.T)
acc = lr.score(x_test.T, y_test.T) * 100

accuracies['Logistic Regression'] = acc
print("Test Accuracy {:.2f}".format(acc))
```

### K-Nearest Neighbour (KNN) Classfication
#### KNN Algorithm


```python
# KNN Model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train.T, y_train.T)
prediction = knn.predict(x_test.T)

print("{} NN Score: {:.2f}".format(2, knn.score(x_test.T, y_test.T) * 100))
```

```python
;# try to find best k value
scoreList = []
for i in range(1, 20):
    knn2 = KNeighborsClassifier(n_neighbors = i)
    knn2.fit(x_train.T, y_train.T)
    scoreList.append(knn2.score(x_test.T, y_test.T))
    
plt.plot(range(1, 20), scoreList)
plt.xticks(np.arange(1, 20, 1))
plt.xlabel("K value")
plt.ylabel("Score")
plt.show()

acc = max(scoreList) * 100
accuracies['KNN'] = acc
print("Maximum KNN Score is {:.2f}".format(acc))
```

#### Support Vector Machine (SVM) Algorithm



```python
from sklearn.svm import SVC
svm = SVC(random_state=1)
svm.fit(x_train.T, y_train.T)

acc = svm.score(x_test.T, y_test.T) * 100
accuracies['SVM'] = acc
print('Test Accuracy of SVM Algorithm: {:.2f}%'.format(acc))
```

#### Naive Bayes Algorithm


```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train.T, y_train.T)
acc = nb.score(x_test.T, y_test.T) * 100
accuracies['Naive Bayes'] = acc
print("Accuracy of Naive Bayes: {:.2f}%".format(acc))
```

#### Decision Tree Algorithm


```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train.T, y_train.T)

acc = dtc.score(x_test.T, y_test.T) * 100
accuracies['Decision Tree'] = acc
print("Decision Tree Test Accuracy {:.2f}".format(acc))
```

#### Random Forest Classification

```python
# Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, random_state=1)
rf.fit(x_train.T, y_train.T)

acc = rf.score(x_test.T, y_test.T) * 100
accuracies['Random Forest'] = acc
print("Random Forest Algorithm Accuracy Score: {:.2f}%".format(acc))
```
