```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('crime.csv', encoding="ISO-8859-1")
df.head()
```

```python
code = pd.read_csv('offense_codes.csv', encoding="ISO-8859-1")
code.head(2)
```

### Things I want to know from this dataset
1. What day/time of the year crimes do most happen? 
2. Where in Boston has the highest crime rate?
3. What kind of crime happen the most?
4. Which district is the busiest and does it match with reporting area?
5. Does the frequency of crimes change over the day?
6. What crime invovles shooting?

```python
# Day
plt.figure(figsize=(7,5))
plt.ylim(37500, 50000)
plt.title('Crime of the Week')
sns.countplot(data=df, x='DAY_OF_WEEK', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
# Hour
plt.figure(figsize=(7,5))
plt.title('Crime of the Time')
sns.countplot(data=df, x='HOUR')
```

```python
# Month
plt.figure(figsize=(7,5))
plt.title('Crime of the Month')
sns.countplot(data=df, x='MONTH')
```

```python
# Day that crime happened
plt.figure(figsize=(7,5))
plt.title('Year that crime happened')
plt.ylim(40000, 110000)
sns.countplot(data=df, x='YEAR')
```

```python
# Reporting Area
plt.figure(figsize=(7,5))
plt.title('Crime Places')
plt.ylim(1000, 2500)
sns.countplot(data=df, x='REPORTING_AREA', order=df['REPORTING_AREA'].value_counts().iloc[1:15].index)
```

```python
# Street that crime happened
plt.figure(figsize=(7,5))
plt.title('Crime Places')
plt.ylim(1000, 2500)
sns.countplot(data=df, y='STREET', order=df['STREET'].value_counts().iloc[1:15].index)
```

```python
# What kind of crime?
plt.figure(figsize=(7,5))
plt.title('Crimes')
plt.xlim(5000, 20000)
sns.countplot(data=df, y='OFFENSE_DESCRIPTION', order=df['OFFENSE_DESCRIPTION'].value_counts().iloc[0:15].index)
```

```python
plt.figure(figsize=(7,5))
plt.title('Crimes')
sns.countplot(data=df, y='OFFENSE_DESCRIPTION', order=df['OFFENSE_DESCRIPTION'].value_counts().iloc[-20:].index)
```

```python
# District Area
plt.figure(figsize=(7,5))
plt.title('District Area')
# plt.ylim(1000, 2500)
sns.countplot(data=df, x='DISTRICT', order=df['DISTRICT'].value_counts().iloc[1:15].index)
```

```python
plt.figure(figsize=(7,5))
plt.title('Crimes')
sns.countplot(data=df, y='OFFENSE_DESCRIPTION', hue='SHOOTING', order=df['OFFENSE_DESCRIPTION'].value_counts().iloc[0:15].index)
```
