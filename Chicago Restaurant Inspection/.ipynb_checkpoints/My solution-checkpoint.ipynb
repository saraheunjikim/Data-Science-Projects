## What do I want to find out?

Soon I'm leaving to Chicago for a week, and I want to find out which restuarnts or market passed the inspection, and which one did not pass, therefore we must avoid. So I want to find out the followings:

1. Find out facility type and which type pass the inspection the most. Restaurant or grocery market?
2. Restaruants with high risk vs. low risk
3. Scatter plot of location of restaurants that passed/didn't pass inspection.
4. find out inspection type, and only find out restaurant that I'm more interested in. 
    (Mostly clean food and facility, rather than violation of tag remover).
    

```python
# Clean out some data

df = df.drop(['Inspection ID', 'AKA Name', 'License #', 'Address', 'City', 'State', 'Inspection Date',
        'Location', 'Historical Wards 2003-2015', 'Zip Codes', 'Community Areas', 'Census Tracts', 'Wards'], axis=1)
```

```python
df.head(2)
```

```python
df.columns = ['Name', 'Type', 'Risk', 'Zip', 'Inspection Type', 'Result', 'Violation', 'Latt', 'Long']
```

```python
df.head(2)
```

```python
df['Type'].describe
```

```python
# Drop rows if it is not Restaurant or Grocery Store.
lst = ['Restaurant', 'Grocery Store']

df = df[df["Type"].isin(lst)]
```

```python
df.head()
```

```python
# Going to make 3 different data frames based on result.
# Restaurants that passed, failed, and passed with conditions.

df_pass = df.loc[(df['Result'] == 'Pass')]
df_fail = df.loc[(df['Result'] == 'Fail')]
df_passwc = df.loc[(df['Result'] == 'Pass w/ Conditions')]
```

```python
df_pass.head(2)
```

```python
df_fail.head(2)
```

```python
df_passwc.head(2)
```

```python
# Basic charts of data frame

plt.figure(figsize=(10, 4))
sns.countplot(data=df, x='Risk')
plt.title('Data Frame')
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df_pass, x='Risk')
plt.title('Data Frame with "Pass" result')
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df_fail, x='Risk')
plt.title('Data Frame with "Fail" result')
```

```python
# Let's scatterplot of passed / failed restaruant and where most of them are located.

plt.figure(figsize=(8, 8))
plt.scatter(data=df_pass, x='Latt', y='Long', color='green', alpha=0.2, s=5)
plt.title('Restaurants that passed inspection test')
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(data=df_fail, x='Latt', y='Long', color='red', alpha=0.2, s=5)
plt.title('Restaurants that failed inspection test')
```

```python
# Let's see what type of inspectin test are mostly taken.

plt.figure(figsize=(10,10))
sns.countplot(data=df, y='Inspection Type', order=df['Inspection Type'].value_counts().iloc[1:20].index)
```

```python
# What kind of inspection test the restaurants passed with?

plt.figure(figsize=(10,10))
sns.countplot(data=df_pass, y='Inspection Type', order=df['Inspection Type'].value_counts().iloc[1:10].index)
```

```python
# What kind of inspection test the restaurants failed with?

plt.figure(figsize=(10,10))
sns.countplot(data=df_fail, y='Inspection Type', order=df['Inspection Type'].value_counts().iloc[1:10].index)
```

```python
# Let's collect only grocery stores and Restaurants

df_grocery = df.loc[(df['Type'] == 'Grocery Store')]
df_restaurant = df.loc[(df['Type'] == 'Restaurant')]
```
