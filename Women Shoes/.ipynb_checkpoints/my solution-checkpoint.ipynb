```python
df = pd.read_csv('women_shoes.csv')
df.head(2)
```

```python
df.columns
```

```python
df.info()
```

```python
df = df.drop(['id', 'dateAdded', 'dateUpdated', 'asins', 'categories', 'primaryCategories', 'ean', 'imageURLs', 'keys',
       'manufacturerNumber', 'prices.sourceURLs', 'upc'], axis=1) 
```

```python
df = df.drop(['dimension', 'prices.dateSeen', 'sourceURLs', 'weight'], axis=1)
```

```python
df.head(2)
```

```python
plt.figure(figsize=(10,50))
plt.xlim(100,500)
sns.countplot(data=df, y='brand')
```

```python
plot_data = df.groupby('brand').size().reset_index(name='count').query('count>=100')
plt.figure(figsize=(20,15))
plt.xlim(100,2000)
sns.barplot(data=plot_data, y='brand', x='count')
```

```python
plot_data.head(2)
```

## What is the average price of each distinct brand listed?

```python
# prices.amountMax

max_price = df[['brand', 'prices.amountMax']].copy()
```

```python
max_price.head(2)
```

```python
max_price.sort_values(by='prices.amountMax', ascending=False)
```

```python
brand_price = max_price.groupby('brand')
```

```python
brand_price
```

```python
for brand, brand_df in brand_price:
    print(brand)
    print(brand_df)
```

```python
brand_price.get_group('Nike').head()
```

```python
brand_price_mean = brand_price.mean()
```

```python
brand_price_mean.head()
```

```python
brand_price_mean = brand_price_mean.sort_values(by='prices.amountMax', ascending=False)
```

```python
brand_price_mean.head()
brand_price_mean = brand_price_mean.reset_index()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=brand_price_mean.loc[:50], y='brand', x='prices.amountMax')
```

```python
plt.figure(figsize=(20,10))
plt.xlim(0,350)
sns.barplot(data=brand_price_mean.loc[80:], y='brand', x='prices.amountMax')
```

## Which ones have the widest distribution of prices?

```python
dist_price = df[['brand','prices.amountMax','prices.amountMin']].copy()
```

```python
dist_price.head()
```

```python
dist_price = dist_price.groupby('brand')
```

```python
for dist, price in dist_price:
    print(dist)
    print(price)
```

```python
dist_price = dist_price.mean()
```

```python
dist_price.head()
```

```python
dist_price['difference'] = dist_price['prices.amountMax'] - dist_price['prices.amountMin']
```

```python
dist_price.head()
```

```python
dist_price = dist_price.sort_values(by='difference', ascending=False)
```

```python
dist_price.head()
```

```python
dist_price['brand'] = dist_price.index
```

```python
plt.figure(figsize=(20,20))
sns.barplot(data=dist_price[:77], y='brand', x='difference')
```
